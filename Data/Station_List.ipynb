```python
import pandas as pd
from datetime import datetime
import os
import numpy as np
import requests
from pandas.io.json import json_normalize
```

### Import and Join Monthly Reports of Bike Trips Taken

```python
url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
response = requests.get(url)
initial = response.json()

station_id=[]
station_name=[]
station_lat=[]
station_lon=[]
station_capacity=[]
station_url=[]
i = 0
for i in range(935):
    station_id.append(initial['data']['stations'][i]['station_id'])
    station_name.append(initial['data']['stations'][i]['name'])
    station_lat.append(initial['data']['stations'][i]['lat'])
    station_lon.append(initial['data']['stations'][i]['lon'])
    station_capacity.append(initial['data']['stations'][i]['capacity'])
    station_url.append(initial['data']['stations'][i]['rental_url'])    
    i +=1


df=  pd.DataFrame(
      {'Station_ID': station_id,
       'Station_Name': station_name,
       'Capacity': station_capacity,
       'URL': station_url,
       'Latitude': station_lat,
       'Longitude': station_lon
      })
df
```

```python
df.to_csv(os.path.join("Clean","Bike_Trip_Histories","station_list.csv"), index = False)
```
