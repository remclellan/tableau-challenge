```python
path = os.path.join("Raw","Monthly_Operating_Reports","2018")
i=0
for WorkingFile in os.listdir(path):
    file = read_pdf(os.path.join(path, WorkingFile),area=[(78.864,117.758,267.965,490.803)],pages="5")[0]
    df = pd.DataFrame(file)
    df[df.columns[1:]] = df[df.columns[1:]].replace('[\D]', '', regex=True).astype(float)
    if i<1:
        df_2018 = df[df.columns[:2]]
    else:
        df = df[df.columns[:2]]
        df_2018 = pd.merge(df_2018, df, on='Revenue Categories')
    i+=1
df_2018
```

```python
path = os.path.join("Raw","Monthly_Operating_Reports","2019")
i=0
for WorkingFile in os.listdir(path):
    file = read_pdf(os.path.join(path, WorkingFile),area=[(78.864,117.758,267.965,490.803)],pages="5")[0]
    df = pd.DataFrame(file)
    df[df.columns[1:]] = df[df.columns[1:]].replace('[\D]', '', regex=True).astype(float)
    if i<1:
        df_2019 = df[df.columns[:2]]
    else:
        df = df[df.columns[:2]]
        df_2019 = pd.merge(df_2019, df, on='Revenue Categories')
    i+=1
df_2019
```

```python
path = os.path.join("Clean","Monthly_Operating_Reports","user_generated")
for WorkingFile in os.listdir(path):
    file = pd.read_csv(os.path.join(path, WorkingFile))
    df = pd.DataFrame(file)
    df[df.columns[1:]] = df[df.columns[1:]].replace('[\D]', '', regex=True).astype(float)
    df = df[df.columns[:2]]
    df_2018 = pd.merge(df_2018, df, on='Revenue Categories')
df_2018
```

```python
df_2018 = df_2018[["Revenue Categories", "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]]
df_2018
```

```python
df_2018.rename(columns = {
    'January':'January 1, 2018',
    'February':'February 1, 2018',
    'March':'March 1, 2018',
    'April':'April 1, 2018',
    'May':'May 1, 2018',
    'June':'June 1, 2018',
    'July':'July 1, 2018',
    'August':'August 1, 2018',
    'September':'September 1, 2018',
    'October':'October 1, 2018',
    'November':'November 1, 2018',
    'December':'December 1, 2018'}, inplace = True)
df_2018
```

```python
df_2019 = df_2019[["Revenue Categories", "January", "February", "March", "April", "May"]]
df_2019
```

```python
df_2019.rename(columns = {
    'January':'January 1, 2019',
    'February':'February 1, 2019',
    'March':'March 1, 2019',
    'April':'April 1, 2019',
    'May':'May 1, 2019'}, inplace = True)
df_2019
```

```python
final_df = pd.merge(df_2018, df_2019, on='Revenue Categories')
final_df
```

```python
final_transposed = final_df.transpose()
final_transposed
```

```python
final_transposed = final_transposed.rename(columns=final_transposed.iloc[0]).drop(final_transposed.index[0])
final_transposed
```

```python
final_transposed.reset_index(inplace=True)
final_transposed
```

```python
final_transposed.rename(columns = {
    'index':'Month'}, inplace = True)
final_transposed
```

```python
final_transposed.dtypes
```

```python
final_transposed[final_transposed.columns[1:]] = final_transposed[final_transposed.columns[1:]].replace('[\D]', '', regex=True).astype(float)
final_transposed
```

```python
final_transposed['Month'] = pd.to_datetime(final_transposed['Month'])
final_transposed.dtypes
```

```python
final_transposed['Year'] = final_transposed['Month'].dt.year
final_transposed
```

```python
final_transposed.to_csv(os.path.join("Clean","Monthly_Operating_Reports","Monthly_Operating_Reports.csv"), index = False)
```
