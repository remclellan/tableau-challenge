### Import and Join Quarterly Reports of Membership and Ridership Data

```python
file = pd.read_csv(os.path.join("Raw","Daily_Ridership_Membership_Data","2019Q2.csv"))
df = pd.DataFrame(file)
df
```

```python
df.drop(['Column 1'], axis=1, inplace = True)
df
```

```python
df.to_csv(os.path.join("Raw","Daily_Ridership_Membership_Data","2019Q2.csv"), index = False)
```

```python
appended_data = []
path = os.path.join("Raw","Daily_Ridership_Membership_Data")
for WorkingFile in os.listdir(path):
    file = pd.read_csv(os.path.join(path, WorkingFile))
    df = pd.DataFrame(file)
    df['Date'] = pd.to_datetime(df['Date'])
    df.rename(columns = {
    'Trips over the past 24-hours (midnight to 11:59pm)':'Trips',
    'Miles traveled today (midnight to 11:59 pm)':'Miles_Traveled',
    'Total Annual Members (All Time)':'Total_Annual_Members',
    '24-Hour Passes Purchased (midnight to 11:59 pm)':'24-Hour_Passes_Purchased',
    '3-Day Passes Purchased (midnight to 11:59 pm)':'3-Day_Passes_Purchased'}, inplace = True)
    df['Year'] = df['Date'].dt.year
    appended_data.append(df)
appended_data = pd.concat(appended_data)
appended_data
```

```python
appended_data.dtypes
```

```python
# def mapper(month):
#     return month.strftime('%b')

# appended_data['Month'] = appended_data['Date'].apply(mapper)
appended_data['Month-Year']= appended_data['Month'].map(str) +'-1-'+ appended_data['Year'].map(str)
appended_data
```

```python
grouped_df=appended_data.groupby('Month-Year').agg({'Date':['nunique'], 
                                                    'Trips': ['sum'], 
                                                    'Miles_Traveled':['sum','mean'],
                                                    '24-Hour_Passes_Purchased':['sum','mean'],
                                                    '3-Day_Passes_Purchased':['sum','mean'],
                                                    'Total_Annual_Members':['min', 'max', 'mean']}).reset_index()
grouped_df
```

```python
grouped_df = grouped_df[grouped_df['Month-Year'] != 'Jun-1-2019']
grouped_df
```

```python
grouped_df['Month-Year'] = pd.to_datetime(grouped_df['Month-Year'])
grouped_df.columns = grouped_df.columns.map('_'.join)
grouped_df
```

```python
grouped_df.to_csv(os.path.join("Clean","Daily_Ridership_Membership_Data","Daily_Ridership_Membership_Data.csv"), index = False)
```
