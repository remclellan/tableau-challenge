### Import and Join Monthly Reports of Bike Trips Taken

```python
file = pd.read_csv(os.path.join("Raw","Bike_Trip_Histories","201801-citibike-tripdata.csv"))
df = pd.DataFrame(file)
df.head()
```

```python
df.dtypes
```

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
df.dtypes
```

```python
df.head()
```

```python
df['gender'] = np.select([df.gender == 1, df.gender == 2], 
                        ['male', 'female'], 
                        default='unknown')
df
```

```python
df['gender'].value_counts()
```

```python
def func(row):
    if row['birth year'] <= 1945:
        return 'Silent'
    elif row['birth year'] <= 1964:
        return 'Boomers' 
    elif row['birth year'] <=1980:
        return 'Generation X'
    elif row['birth year'] <=1996:
        return 'Millennials'
    else:
        return 'Generation Z'

df['generation'] = df.apply(func, axis=1)
df.head()
```

```python
df.rename(columns = {
    'tripduration':'duration (min)',
    'starttime':'start',
    'stoptime':'stop',}, inplace = True)
df.drop(['start station name', 'start station latitude','start station longitude','end station name','end station latitude','end station longitude'],axis=1, inplace=True)
df
```

```python
appended_data = []
path = os.path.join("Raw","Bike_Trip_Histories","2018Q1")
for WorkingFile in os.listdir(path):
    file = pd.read_csv(os.path.join(path, WorkingFile))
    df = pd.DataFrame(file)
    df['starttime'] = pd.to_datetime(df['starttime'])
    df['stoptime'] = pd.to_datetime(df['stoptime'])
    df.rename(columns = {
    'tripduration':'duration (min)',
    'starttime':'start',
    'stoptime':'stop',}, inplace = True)
    df['gender'] = np.select([df.gender == 1, df.gender == 2], 
                        ['male', 'female'], 
                        default='unknown')
    def func(row):
        if row['birth year'] <= 1945:
            return 'Silent'
        elif row['birth year'] <= 1964:
            return 'Boomers' 
        elif row['birth year'] <=1980:
            return 'Generation X'
        elif row['birth year'] <=1996:
            return 'Millennials'
        else:
            return 'Generation Z'

    df['generation'] = df.apply(func, axis=1)

    appended_data.append(df)
Q1_2018 = pd.concat(appended_data)
Q1_2018.head()
```

```python
def mapper(month):
    return month.strftime('%b')

Q1_2018['Month'] = Q1_2018['start'].apply(mapper)
Q2_2018['Month'] = Q2_2018['start'].apply(mapper)
Q3_2018['Month'] = Q3_2018['start'].apply(mapper)
Q4_2018['Month'] = Q4_2018['start'].apply(mapper)
Q1_2019['Month'] = Q1_2019['start'].apply(mapper)
Q2_2019['Month'] = Q2_2019['start'].apply(mapper)
```

```python
Q1_2018['age'] = Q1_2018['Year'] - Q1_2018['birth year']
Q2_2018['age'] = Q2_2018['Year'] - Q2_2018['birth year']
Q3_2018['age'] = Q3_2018['Year'] - Q3_2018['birth year']
Q4_2018['age'] = Q4_2018['Year'] - Q4_2018['birth year']
Q1_2019['age'] = Q1_2019['Year'] - Q1_2019['birth year']
Q2_2019['age'] = Q2_2019['Year'] - Q2_2019['birth year']
```

```python
Q1_2018['Month-Year']= Q1_2018['Month'].map(str) +'-1-'+ Q1_2018['Year'].map(str)
Q2_2018['Month-Year']= Q2_2018['Month'].map(str) +'-1-'+ Q2_2018['Year'].map(str)
Q3_2018['Month-Year']= Q3_2018['Month'].map(str) +'-1-'+ Q3_2018['Year'].map(str)
Q4_2018['Month-Year']= Q4_2018['Month'].map(str) +'-1-'+ Q4_2018['Year'].map(str)
Q1_2019['Month-Year']= Q1_2019['Month'].map(str) +'-1-'+ Q1_2019['Year'].map(str)
Q2_2019['Month-Year']= Q2_2019['Month'].map(str) +'-1-'+ Q2_2019['Year'].map(str)
```

```python
Q1_2018.head()
```

```python
Q1_2018.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
Q2_2018.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
Q3_2018.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
Q4_2018.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
Q1_2019.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
Q2_2019.drop(['start', 'stop', 'birth year', 'Month', 'Year'],axis=1, inplace=True)
```

```python
grouped_Q12018=Q1_2018.groupby(['Month-Year','start station name','end station name','generation','gender','usertype']).agg({'bikeid':['count'],
                                                        'duration (min)': ['sum', 'mean', 'max', 'min'],
                                                        'age':['mean']}).reset_index()
grouped_Q12018
```
